```
import pandas as pd
import numpy as np
import requests
from requests.exceptions import HTTPError
import pickle
from time import sleep, time
```

```
PATH = "/content/drive/My Drive/Datasets/Movies.csv"
df = pd.read_csv(PATH)
df = df[['movieId', 'tmdbId']]
df.head()
```

```
def scrape_actors():
    api_key='5791e90fa4bae4661acf8540e91c7225'
    results = []
    for index, row in df.iterrows():
        if index%100==0: print(index)
        tmdbId = row['tmdbId']
        url = f'https://api.themoviedb.org/3/movie/{tmdbId}/credits?api_key={api_key}'

        try:
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()
            curr_res = {
                'tmdbId': tmdbId,
                'actors': []
            }
            for actor in data['cast'][:3]:
                curr_res['actors'].append({'name': actor['name'], 'profilePath': actor['profile_path']})
            results.append(curr_res)
            with open('results', 'wb') as fi:
                pickle.dump(results, fi)
            sleep(0.4)
        except HTTPError as http_err:
            print(f'HTTP error occurred: {http_err} at url: {url}')
        except Exception as e:
            print("An error occurred: ", e, " at url: ", url)

start = time()
scrape_actors()
print("Time taken: ", time()-start)
```

```
with open('results', 'rb') as fi:
    results = pickle.load(fi)
    print(results)
```

```
def scrape_poster():
    api_key='5791e90fa4bae4661acf8540e91c7225'
    results = []
    for index, row in df.iterrows():  
        try:
            if index%100==0: print(index)
            tmdbId = row['tmdbId']
            url = f'https://api.themoviedb.org/3/movie/{tmdbId}/images?api_key={api_key}'
            tmdbId = row['tmdbId']
    
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()
            
            poster_url_path = data['posters'][0]['file_path']
            results.append({'tmdbId': tmdbId, 'posterUrl': poster_url_path})
            with open('results', 'wb') as fi:
                pickle.dump(results, fi)
            sleep(0.4)
        except HTTPError as http_err:
            print(f'HTTP error occurred: {http_err} at url: {url}')
        except Exception as e:
            print("An error occurred: ", e, " at url: ", url)
start = time()
scrape_poster()
print("Time taken: ", time()-start)
```
